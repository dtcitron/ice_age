```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plot Ratings over Time

```python
plt.figure(figsize=(8,6))

x = np.array([1, 2, 3, 4, 5])
# Ratings - Rotten Tomatoes and Metacritic
rt_c = np.array([77, 57, 45, 37, 11]) # Rotten Tomatoes, Critics
rt_a = np.array([82, 71, 62, 62, 40]) # Rotten Tomatoes, Audiences
mc_c = np.array([60, 58, 50, 49, 34]) # Metacritics, Critics
mc_a = np.array([8, 7.6, 6.8, 5.8, 4.1])*10 # Metacritic, Audiences

# plot the data
plt.plot(x, rt_c, marker = 'o', linewidth = 3)
plt.plot(x, rt_a, marker = 'o', linewidth = 3)
plt.plot(x, mc_c, marker = 'o', linewidth = 3)
plt.plot(x, mc_a, marker = 'o', linewidth = 3)
plt.xlim(.5,5.5)
plt.ylim(0,100)

# setting ticks
# http://matplotlib.org/examples/ticks_and_spines/ticklabels_demo_rotation.html
labels = [r'Ice Age', r'Ice Age 2', r'Ice Age 3', r'Ice Age 4', r'Ice Age 5']
plt.xticks(x, labels, rotation=45, fontsize=18)
labels = [0, 20, 40, 60, 80, 100]
plt.yticks(labels, labels, fontsize=18)

# setting legend
plt.legend(["Rotten Tomatoes - Critics\' Scores", "Rotten Tomatoes - Audience Scores", 
           "Metacritic - Critics\' Scores", "Metacritic - Audience Scores"], loc='lower left', fontsize=16);
plt.savefig('theratings.pdf', bbox_inches = 'tight');
```

## Plot Domestic Opening Weekend and Global Gross

```python
plt.figure(figsize=(8,6))
x1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([1, 2, 3, 4])

o_us= np.array([ 46.312454,  68.033544,  41.690382,  46.629259, 21.3]) # Domestic opening
d = np.array([ 176.387405,  195.330621,  196.573705,  161.321843]) # Domestic gross
w = np.array([ 383.257136,  660.94078 ,  690.113112,  715.922939]) # Worldwide gross

# Multiple y-axes with shared x-axis
# http://matplotlib.org/examples/subplots_axes_and_figures/fahrenheit_celsius_scales.html
# http://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib
fig, ax1 = plt.subplots(figsize=(8,6))
ax2 = ax1.twinx()

ax1.plot(x1, o_us, marker = 'o', color = 'k', linewidth=3)
ax1.set_ylabel('US Opening Weekend \n (Millions USD)', color = 'k', fontsize = 18)

labels = [r'Ice Age', r'Ice Age 2', r'Ice Age 3', r'Ice Age 4', r'Ice Age 5']
ax1.set_xticks(x1)
ax1.set_xlim(.5,5.5)
ax1.set_xticklabels(labels, rotation=45, fontsize=18)
ax1.set_yticks([20, 30, 40, 50, 60, 70])
ax1.set_yticklabels([20, 30, 40, 50, 60, 70], fontsize=18)

ax2.plot(x2, d+w, marker = 'o', color = 'r', linewidth=3)
ax2.set_ylabel('Worldwide Gross \n (Millions USD)', color = 'r', fontsize = 18)
ax2.set_yticks([550, 600, 650, 700, 750, 800, 850, 900])
ax2.set_yticklabels([550, 600, 650, 700, 750, 800, 850, 900], fontsize=18)
#plt.savefig('themoney.pdf', bbox_inches = 'tight');
```
